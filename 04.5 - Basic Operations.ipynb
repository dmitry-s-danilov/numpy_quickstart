Many unary operations,
such as computing the **sum**
of all the elements in the array,
are implemented as methods
of the ndarray class.
By default,
these operations apply to the array
as though it were a list of numbers,
regardless of its shape.
However, by specifying the **axis** parameter
you can apply an operation
along the specified axis of an array.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# apply min, max, sum and cumsum methods
# to the array
# and along the specified axis 

n, m = 3, 4
x = np.arange(1, n * m + 1).reshape((n, m))

print(
    f'{type(x)}:', x,
    f'min: {x.min()}',
    f'max: {x.max()}',
    f'sum: {x.sum()}',
    sep='\n', end='\n\n')

for i in range(x.ndim):
    print(
        f'axis: {i}',
        f'min: {x.min(axis=i)}',
        f'max: {x.max(axis=i)}',
        f'sum: {x.sum(axis=i)}',
        f'cumsum:', x.cumsum(axis=i),
        sep='\n', end='\n\n')
```

```python
# apply mean, min and max methods
# to the array
# and along axis-0

k, n = 2, 10 + 1 
x = np.random.random((k, n))

for i in range(k):
    plt.plot(x[i], '.:', label=f'random')
    plt.plot(
        [0, x.shape[1] - 1],
        [x[i].mean(), x[i].mean()], 
        '-',
        label='mean: {:.3f}'.format(x[i].mean()))
    plt.plot(
        x[i].argmin(),
        x[i].min(),
        'o',
        label='min: {:.3f}'.format(x[i].min()))
    plt.plot(
        x[i].argmax(),
        x[i].max(),
        'o',
        label='max: {:.3f}'.format(x[i].max()))
    plt.xlabel('index')
    plt.ylabel('value')
    plt.title(f'random: {i}')
    plt.legend()
    plt.show()

for i in range(k):
    plt.plot(x[i], '.:', label=f'random {i}')
plt.plot(
        [0, x.shape[1] - 1],
        [x.mean(), x.mean()], 
        '-',
        label='mean {:.3}'.format(x.mean()))
plt.plot(
    x.argmin() % x.shape[1],
    x.min(),
    'o',
    label='min: {:.3f}'.format(x.min()))
plt.plot(
    x.argmax() % x.shape[1],
    x.max(),
    'o',
    label='max: {:.3f}'.format(x.max()))
plt.xlabel('index')
plt.ylabel('value')
plt.title('random')
plt.legend()
plt.show()    
```
