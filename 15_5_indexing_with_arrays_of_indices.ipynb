```python
import numpy as np
import matplotlib.pyplot as plt
```

### Time-dependent series

```python
# Creates 1D array of growing sequence,
# which is actually time scale.
t_start, t_stop, t_num = 0, 1, 10 + 1
t = np.linspace(start=t_start, stop=t_stop,
                num=t_num)
# Creates 2D array of randoms,
# which is actually time-depended series.
x_shape = t_num, 3
x = np.random.randn(*x_shape)
```

```python
# Prints time-dependent series.
print('time:', t, t.shape, '',
      'series:', x, x.shape,
      sep='\n', end='\n\n')
```

```python
# Plots time-dependent series.

plt.axhline(y=0, lw=1, c='k')

for j in range(x.shape[1]):
    plt.plot(t, x[:, j], 'o:', label=f'series {j}')

plt.xlabel('time')
plt.title('standard normal')
plt.legend()

plt.show()
```

### Statistics of series

```python
# Finds statistics of series
# without reference to time scale.

# Finds minima, using method,
# for each series.
x_min = x.min(axis=0)
# The same can be obtained
# using function.
# x_min = np.min(x, axis=0)

# Finds maxima, using method,
# for each series.
x_max = x.max(axis=0)
# The same can be obtained
# using function.
# x_max = np.max(x, axis=0)

# Finds mean, using method,
# for each series.
x_mean = x.mean(axis=0)
# The same can be obtained
# using function.
# x_mean = np.mean(x, axis=0)

# Finds median, using function,
# for each series.
x_median = np.median(x, axis=0)

print('series:', x,
      'mix:', x_min,
      'max:', x_max,
      'mean:', x_mean,
      'median:', x_median,
      sep='\n')
```

```python
# Prints time-dependent series
# together with statistics.

# Prints title.
print('{:>6}'.format('t'), end=' ')
for j in range(x.shape[1]):
        print(f'{j:>6}', end=' ')
print()

# Prints series.
for i in range(x.shape[0]):
    print(f'{t[i]:6.1f}', end=' ')
    for j in range(x.shape[1]):
        print(f'{x[i, j]:>6.3f}', end=' ')
    print()

# Prints min.
print('{:>6}'.format('min'), end=' ')
for _ in x_min:
        print(f'{_:>6.3f}', end=' ')
print()
    
# Prints max.
print('{:>6}'.format('max'), end=' ')
for _ in x_max:
        print(f'{_:>6.3f}', end=' ')
print()

# Prints mean.
print('{:>6}'.format('mean'), end=' ')
for _ in x_mean:
        print(f'{_:>6.3f}', end=' ')
print()

# Prints median.
print('{:>6}'.format('median'), end=' ')
for _ in x_median:
        print(f'{_:>6.3f}', end=' ')
```

```python
# Plots time-dependent series
# together with statistics.

fig_x_size, fig_y_size = 6.4, 3.2
fig_size = [fig_x_size, x.shape[1] * fig_y_size]
fig, ax = plt.subplots(nrows=x.shape[1], ncols=1,
                       sharex=True, figsize=fig_size)

for j in range(x.shape[1]):
    ax[j].axhline(y=0, lw=1, c='k')
    ax[j].plot(t, x[:, j], 'o:',
               label=f'series: {j}')
    ax[j].plot((t[0], t[-1]),
               (x_mean[j], x_mean[j]), '-',
               label=f'mean: {x_mean[j]:6.3f}')
    ax[j].plot((t[0], t[-1]),
               (x_median[j], x_median[j]), '-',
               label=f'median: {x_median[j]:6.3f}')
    ax[j].legend()

ax[-1].set_xlabel('time')
fig.suptitle('standard normal')
plt.show()
```

### Indeces of minima and maxima

```python
# Finds index of the minima
# for each series.
i_min = x.argmin(axis=0)
# Finds times corresponding
# to the minima.
t_min = t[i_min]
# Finds minima values of series.
x_min_1 = x[i_min, range(x.shape[1])]

# Finds index of the maxima
# for each series.
i_max = x.argmax(axis=0)
# Finds times corresponding
# to the maxima
t_max = t[i_max]
# Finds maxima values of series.
x_max_1 = x[i_max, range(x.shape[1])]
```

```python
print('min',
      'i:', i_min,
      't:', t_min,
      'x:', x_min_1,
      'check:', np.all(x_min == x_min_1), '',
      
      'max',
      'i:', i_max,
      't:', t_max,
      'x:', x_max_1,
      'check:', np.all(x_max == x_max_1),
      sep='\n')
```

```python
# Plots time-dependent series
# together with minima and maxima.

fig_x_size, fig_y_size = 6.4, 3.2
fig_size = [fig_x_size, x.shape[1] * fig_y_size]
fig, ax = plt.subplots(nrows=x.shape[1], ncols=1,
                       sharex=True, figsize=fig_size)

for j in range(x.shape[1]):
    ax[j].plot(t, x[:,j], 'o:',
               label=f'series: {j}')
    ax[j].plot(t_min[j], x_min[j], 'o',
               label=f'min: ({t_min[j]:3.1f}, {x_min[j]:6.3f})')
    ax[j].plot(t_max[j], x_max[j], 'o',
               label=f'max: ({t_max[j]:3.1f}, {x_max[j]:6.3f})')
    ax[j].axhline(y=0, lw=1, c='k')
    ax[j].legend()

    fig.suptitle('stabdard normal')
plt.show()
```
