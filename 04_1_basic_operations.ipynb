# The basics

## Basic operations

### Elementwise operations

- Arithmetic operators on arrays apply elementwise.
- A new array is created and filled with the result.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
title="""add a number to an array
subtract a number from an array"""

x = np.random.randn(11)
y = 1

plt.plot(x, '.:', label='x')
plt.plot(x + y, 'o-', label=f'x + {y}')
plt.plot(x - y, 'o-', label=f'x - {y}')

# plt.axhline(y=0, linewidth=1, color='k')
plt.axhline(y=0, lw=1, c='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title(title)
plt.legend()
plt.show()
```

```python
title="""add an array to an array
subtract an array from an array"""

x, y = np.random.randn(2, 11)

plt.plot(x, '.:', label='x')
plt.plot(y, '.:', label='y')
plt.plot(x + y, 'o-', label='x + y')
plt.plot(x - y, 'o-', label='x - y')

plt.axhline(y=0, lw=1, c='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title(title)
plt.legend()
plt.show()
```

```python
title="""multiply an array by a number
divide an array by a number"""

x = np.random.randn(11)
y = 2

plt.plot(x, '.:', label='x')
plt.plot(x * y, 'o-', label=f'x * {y}')
plt.plot(x / y, 'o-', label=f'x / {y}')

plt.axhline(y=0, lw=1, c='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title(title)
plt.legend()
plt.show()
```

```python
title="""multiply an array by an array
divide an array by an array"""

x, y = np.random.rand(2, 11)

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

ax[0].plot(x, '.:', label='x')
ax[0].plot(y, '.:', label='y')
ax[0].plot(x * y, 'o-', label='x * y')
ax[0].set_ylabel('value')
ax[0].set_title(title.split('\n')[0])
ax[0].legend()

ax[1].plot(x, '.:', label='x')
ax[1].plot(y, '.:', label='y')
ax[1].plot(x / y, 'o-', label='x / y')
ax[1].set_ylabel('value')
ax[1].set_xlabel('index')
ax[1].set_title(title.split('\n')[1])
ax[1].legend()

fig.set_size_inches(8, 6)
fig.subplots_adjust(hspace=.3)

plt.show()
```

```python
title='raise an array to the power of a number'

x = np.random.randn(11)
y = 2

plt.plot(x, '.:', label='x')
plt.plot(x ** y, 'o-', label=f'x ** {y}')

plt.axhline(y=0, lw=1, c='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title(title)
plt.legend()
plt.show()
```

```python
title='apply a function to an array'

x = np.random.randn(11)

plt.plot(x, '.:', label='x')
plt.plot(np.exp(x), 'o-', label='exp(x)')

plt.axhline(y=0, lw=1, c='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title(title)
plt.legend()
plt.show()
```

```python
title='apply conditions to an array'

x = np.random.rand(11)
x_indexes = np.arange(x.size)

x_median = np.median(x)

x_median_equal = x == x_median
x_median_equal_indexes = x_indexes[x_median_equal]
x_median_equal_values = x[x_median_equal]

x_median_less = x < x_median
x_median_less_indexes = x_indexes[x_median_less]
x_median_less_values = x[x_median_less]

x_median_more = x > x_median
x_median_more_indexes = x_indexes[x_median_more]
x_median_more_values = x[x_median_more]

print(f'array:\n{x}',
      f'median: {x_median}',
      f'median indexes: {x_median_equal_indexes}',
      f'median less indexes: {x_median_less_indexes}',
      f'median more indexes: {x_median_more_indexes}',
      f'median less values:\n{x_median_less_values}',
      f'median more values:\n{x_median_more_values}',
      sep='\n')

fig, ax = plt.subplots(nrows=2, ncols=1,
                       sharex=True, sharey=True)

ax[0].plot(x, 'o:', label='random')
ax[0].plot((0, x.size - 1), (x_median, x_median),
           '--', lw=1, c='k', label='median')
ax[0].set_ylabel('value')
ax[0].legend()

ax[1].plot(x, ':', label='random')
ax[1].plot(x_median_equal_indexes, x_median_equal_values,
           'o', label='median')
ax[1].plot(x_median_less_indexes, x_median_less_values,
           'o', label='median less')
ax[1].plot(x_median_more_indexes, x_median_more_values,
           'o', label='median more')
ax[1].plot((0, x.size - 1), (x_median, x_median),
           '--', lw=1, c='k', label='median')
ax[1].set_xlabel('index')
ax[1].set_ylabel('value')
ax[1].legend()

fig.set_size_inches(8, 6)
fig.subplots_adjust(hspace=.3)
fig.suptitle(title)

plt.xlim(0, x.size - 1)
plt.show()
```
