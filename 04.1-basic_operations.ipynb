Arithmetic operators on arrays apply elementwise.
A new array is created and filled with the result.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# add a number to an array
# subtract a number from an array

n = 10 + 1
x = np.random.randn(n)
y = 1

plt.plot(x, '.:', label='x')
plt.plot(x + y, 'o-', label=f'x + {y}')
plt.plot(x - y, 'o-', label=f'x - {y}')
plt.axhline(y=0, linewidth=1, color='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# add an array to an array
# subtract an array from an array

n = 10 + 1
z = np.random.randn(2, n)
x, y = z

plt.plot(x, '.:', label='x')
plt.plot(y, '.:', label='y')
plt.plot(x + y, 'o-', label='x + y')
plt.plot(x - y, 'o-', label='x - y')
plt.axhline(y=0, linewidth=1, color='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# multiply and divide
# an array by a number

n = 10 + 1
x = np.random.randn(n)
y = 2

plt.plot(x, '.:', label='x')
plt.plot(x * y, 'o-', label=f'x * {y}')
plt.plot(x / y, 'o-', label=f'x / {y}')
plt.axhline(y=0, linewidth=1, color='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# multiply and divide
# an array by an array

n = 10 + 1
z = np.random.rand(2, n)
x, y = z

plt.plot(x, '.:', label='x')
plt.plot(y, '.:', label='y')
plt.plot(x * y, 'o-', label='x * y')
plt.xlabel('index')
plt.ylabel('value')
plt.title('random')
plt.legend()
plt.show()

plt.plot(x, '.:', label='x')
plt.plot(y, '.:', label='y')
plt.plot(x / y, 'o-', label='x / y')
plt.xlabel('index')
plt.ylabel('value')
plt.title('random')
plt.legend()
plt.show()
```

```python
# exponentiate an array

n = 10 + 1
x = np.random.randn(n)
y = 2

plt.plot(x, '.:', label='x')
plt.plot(x ** y, 'o-', label=f'x ** {y}')
plt.axhline(y=0, linewidth=1, color='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# apply a function to an array

n = 10 + 1
x = np.random.randn(n)
y = np.exp(x)

plt.plot(x, '.:', label='x')
plt.plot(y, 'o-', label='exp(x)')
plt.axhline(y=0, linewidth=1, color='k')
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()
```

```python
# apply a condition to an array

n = 10 + 1
x = np.random.randn(n)
m = np.median(x) 
y = x[x < m]
z = x[x > m] 
      
plt.plot(
    x,
    'o:', label='random')
plt.plot(
    (0, x.size - 1), [m, m],
    '-', label='median')
plt.axhline(
    y=0,
    linewidth=1, color='k')
plt.xlim(0, x.size - 1)
plt.xlabel('index')
plt.ylabel('value')
plt.title('standard normal')
plt.legend()
plt.show()

plt.plot(
    y,
    'o:', label='less')
plt.plot(
    z,
    'o:', label='more')
plt.plot(
    (0, y.size - 1), [m, m],
    '-', label='median')
plt.axhline(
    y=0,
    linewidth=1, color='k')
plt.xlim(0, y.size - 1)
plt.xlabel('index')
plt.ylabel('value')
plt.title('random')
plt.legend()
plt.show()
```
