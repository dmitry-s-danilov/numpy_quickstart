# Universal Functions

* NumPy provides familiar mathematical functions
  such as **sin**, **cos**, and **exp**.
* In NumPy, these are called “universal functions” (ufunc).
  Within NumPy, these functions operate elementwise on an array,
  producing an array as output.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Split the interval.

x_min, x_max = -2, 2
n = 50
x = np.linspace(start=x_min, stop=x_max, num=2 * n + 1)

# Plot the functions.

k_x = 2
x_ticks = np.linspace(start=x_min, stop=x_max, num=2 * k_x + 1)

y_min, y_max = -1, 1
k_y = 2
y_ticks = np.linspace(start=y_min, stop=y_max, num=2 * k_y + 1)

plt.plot(x, np.sin(np.pi * x / 2), label='sin')
plt.plot(x, np.cos(np.pi * x / 2), label='cos')

plt.xlim(xmin=x_min, xmax=x_max)
plt.ylim(ymin=y_min, ymax=y_max)

plt.xticks(x_ticks)
plt.yticks(y_ticks)

plt.vlines(x_ticks[1: -1], ymin=y_min, ymax=y_max,
           linewidth=.5, color='grey')
plt.hlines(y_ticks[1: -1], xmin=x_min, xmax=x_max,
           linewidth=.5, color='grey')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
# Split the interval.

x_min, x_max = -2, 2
n = 50
x = np.linspace(start=x_min, stop=x_max, num=2 * n + 1)

# Plot the functions.

k_x = 2
x_ticks = np.linspace(start=x_min, stop=x_max, num=2 * k_x + 1)

y_min, y_max, y_step = -1, 1.5, .5 
y_ticks = np.arange(start=y_min, stop=y_max + y_step, step=y_step)

plt.plot(x, np.exp(-x ** 2) * np.sin(np.pi * x / 2),
         label='sin')
plt.plot(x, np.exp(-x ** 2) * np.cos(np.pi * x / 2),
         label='cos')

plt.xlim(xmin=x_min, xmax=x_max)
plt.ylim(ymin=y_min, ymax=y_max)

plt.xticks(x_ticks)
plt.yticks(y_ticks)

plt.vlines(x_ticks[1: -1], ymin=y_min, ymax=y_max,
           linewidth=.5, color='grey')
plt.hlines(y_ticks[1: -1], xmin=x_min, xmax=x_max,
           linewidth=.5, color='grey')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
# Split the interval.

x_min, x_max = -8, 8
n = 50
x = np.linspace(start=x_min, stop=x_max, num=2 * n)

# Plot the functions.

k_x = 2
x_ticks = np.linspace(start=x_min, stop=x_max, num=2 * k_x + 1)

y_min, y_max, y_step = -.5, 1.5, .5 
y_ticks = np.arange(start=y_min, stop=y_max + y_step, step=y_step)

plt.plot(x, np.sin(np.pi * x / 2) / (np.pi * x / 2))

plt.xlim(xmin=x_min, xmax=x_max)
plt.ylim(ymin=y_min, ymax=y_max)

plt.xticks(x_ticks)
plt.yticks(y_ticks)

plt.vlines(x_ticks[1: -1], ymin=y_min, ymax=y_max,
           linewidth=.5, color='grey')
plt.hlines(y_ticks[1: -1], xmin=x_min, xmax=x_max,
           linewidth=.5, color='grey')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
